```python
def count_words(description, words):
    """
    Counts the number of times any word from a list of words occurs in a description.
    The matching is case-insensitive and matches whole words only.

    Args:
        description (str): The text description to search within.
        words (list): A list of words to count.

    Returns:
        int: The total number of times any of the words in the list occurs 
             in the description.
    """

    count = 0
    description = description.lower()  # Case-insensitive matching

    for word in words:
        # Create a regular expression to match whole words only.
        # \b matches word boundaries.
        pattern = r"\b" + re.escape(word.lower()) + r"\b"  
        matches = re.findall(pattern, description)
        count += len(matches)

    return count
```

```python
def add_space_before_capital(string):
    # Use regex to add space before a capital letter
    result = ""
    if isinstance(string, str):
        result = re.sub(r'(?<!^)(?=[A-Z])', ' ', string)
    return result
    
def remove_punctuation(text):
    print("inside remove_punctuation function")
    if isinstance(text, str):
        return re.sub(r'[^\w\s]', ' ', text)
    return ""
def replace_newline_and_slash(text):
    if isinstance(text, str):
        return text.replace('\n', ' ').replace('/', ' ')
    return ""
```

```python
lemmatizer = WordNetLemmatizer()
# Use defaultdict to handle missing weights gracefully
feminine_weights = defaultdict(int)
for word, weight in feminine_words.items():
    feminine_weights[lemmatizer.lemmatize(word)] = weight

masculine_weights = defaultdict(int)
for word, weight in masculine_words.items():
    masculine_weights[lemmatizer.lemmatize(word)] = weight
```

```python
def fem_words_found(row):
    description = remove_punctuation(row['description'])
    description = replace_newline_and_slash(description)
    description = str(description).lower()
    description_words = [lemmatizer.lemmatize(word) for word in description.split()]
    fem_words_found = []
    for word in description_words:
        if feminine_weights[word]!=0:
            fem_words_found.append(word)
    return fem_words_found

def masc_words_found(row):
    description = remove_punctuation(row['description'])
    description = replace_newline_and_slash(description)
    description = str(description).lower()
    description_words = [lemmatizer.lemmatize(word) for word in description.split()]
    masc_words_found = []
    for word in description_words:
        if masculine_weights[word]!=0:
            masc_words_found.append(word)
    return masc_words_found


def calculate_score(row):
        print("starting now")
        description = remove_punctuation(row['description'])
        print("removed punctuation")
        # description = add_space_before_capital(description)
        description = replace_newline_and_slash(description)
        print("newline removed")
        description = str(description).lower()
        print("lowered")
        description_words = [lemmatizer.lemmatize(word) for word in description.split()]
        # print(description_words)
        weighted_pull = 0
        weighted_push = 0
        fem_words_found, masc_words_found = [], []
        for word in description_words:
            if feminine_weights[word]!=0:
                fem_words_found.append(word)
            if masculine_weights[word]!=0:
                masc_words_found.append(word)
            weighted_pull += feminine_weights[word]
            weighted_push += masculine_weights[word]
        N = weighted_pull + weighted_push

        if N == 0:
            return 0.5

        x = (1 / N) * (weighted_pull - weighted_push)
        score = 1 / (1 + np.exp(-x))
        return score

df['score'] = df.apply(calculate_score, axis=1)
df.head()
```

```python
def contains_diversity_keywords(text):
    text = text.lower()
    if "equal employment opportunit" in text or "equal employments opportunit" in text or "equal opportunit" in text or "affirmative action" in text:
        return 1
    keywords = [
        "diversity", "inclusion", "inclusive", "discriminate", "discrimination",
        "race", "color", "religion", "gender", "gender identity", "gender expression", "sexual orientation", "national origin", "genetics", 
        "disability", "age", "veteran status", "marital status", "pregnancy", "civil status", "family status", "education level"
    ]
    count = sum(1 for keyword in keywords if re.search(rf'\b{re.escape(keyword)}\b', text))
    
    return 1 if count >= 3 else 0
```

```python
## check if equal opportunity employer
df['equal_opportunity_employer'] = df['description'].apply(lambda x: contains_diversity_keywords(x))

print(df["equal_opportunity_employer"].value_counts())
```

```python
## count of male of male, female, gender neutral pronouns used
masc_pronouns = ["he", "him", "his", "himself"]
fem_pronouns = ["she", "her", "hers", "herself"]
neutral_pronouns = ["they", "them", "their", "theirs", "themselves"]

# Function to count occurrences of pronouns in a text
def count_pronouns(text, pronoun_list):
    if pd.isna(text):  # Handle NaN values
        return 0
    description = remove_punctuation(text)
    description = replace_newline_and_slash(description)
    description = str(description).lower()
    return sum(len(re.findall(rf'\b{pronoun}\b', text)) for pronoun in pronoun_list)
```

```python
# Apply the function to each row in the "description" column
df["masc_pronouns_count"] = df['description'].apply(lambda x: count_pronouns(x, masc_pronouns), axis=1)
df["fem_pronouns_count"] = df["description"].apply(lambda x: count_pronouns(x, fem_pronouns), axis=1)
df["neutral_pronouns_count"] = df["description"].apply(lambda x: count_pronouns(x, neutral_pronouns), axis=1)

print(df["masc_pronouns_count"].value_counts())
print(df["fem_pronouns_count"].value_counts())
print(df["neutral_pronouns_count"].value_counts())
df.head()
```
